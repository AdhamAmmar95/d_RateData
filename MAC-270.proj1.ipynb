Cause of death reported in the dataset is obtained from the NYC death certificate which is issued for every death that happens in New York City. The dataset consists of 1,097 rows and 7 columns. The causes of death listed in the dataset varies widely from causes related to a disease such as diabetes and HIV to causes not related to diseases such as homicide and accidents.

# Data Dictionary - Column Information	

		

* **Year**-                      The year of death.(2007-2016)


* **Leading Cause**-	        The cause of death.(Top 10 leading causes by each category.)


* **Sex**-	                    The decendent's sex.(Male, Female.)


* **Race Ethnicity**-	        The decendent's ethnicity.(Hispanic, Asian and Pacific Islancer, Non-Hispanic White, Non-                                        Hispanic Black, Other Race/Ethnicity, Not Stated/Unknown.)

* **Deaths**-	                The number of people who died due to cause of death.(Counts under 5 are suppressed.)


* **Death Rate**-               The death rate within the sex and Race/ethnicity category.(Death rate per 100,000                                                population.Rates with relative standard error (RSE) >30 are suppressed.)

* **Age Adjusted Death Rate**-	The age-adjusted death rate within the sex and Race/ethnicity category.(Age-adjusted death rate                                  per 100,000 population.Rates with relative standard error (RSE) >30 are suppressed.)
